Importing all the libraries that I will be using to analyze the statistics.

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved 
```

Placing the csv file into a Pandas DataFrame.
Writing a for loop to view all of the columns within said DataFrame.

```python
df = pd.read_csv('/kaggle/input/covid19-in-usa/us_states_covid19_daily.csv')
for col in df:
    print(col)
df
```

```python
df_hi_states = df[df.positive >= 1000000]
df_hi_states
```

```python
plt.figure(figsize=(8, 9))
plt.bar(df_hi_states['state'], df_hi_states['positive'], color='red')
plt.title('Three States With Highest COVID-19 Cases')
plt.xlabel('Top 3 States')
plt.ylabel('Amount of Cases in the Millions')
```

Filtering the "state" column to take any data that contains "CA" (California).

```python
df_CA = df[df.state == 'CA']
df_CA.head()
```

Sorting the Dataframe by 'date' in ascending order.

```python
df_CA.sort_values(by='date', inplace=True)
df_CA
```

Plotting on a line graph the amount of positive cases in California

```python
plt.plot(df_CA['date'], df_CA['positive'], color='red')
plt.title('COVID-19 Positive Cases in California')
plt.ylabel('Number of Cases in Millions')
plt.xlabel('From 03-04-20 - 12-06-20')
plt.xticks([])
plt.show()
```

Plotting the total number of deaths in California.

```python
plt.plot(df_CA['date'], df_CA['death'], color='black')
plt.title('COVID-19 Deaths in California')
plt.ylabel('Number of Deaths')
plt.xlabel('From 01-22-20 - 12-06-20')
plt.xticks([])
plt.show()
```

Plotting both lines on one graph to show the contrast.

```python
plt.plot(df_CA['date'], df_CA['positive'], color='red', label='Positive')
plt.plot(df_CA['date'], df_CA['death'], color ='black', label='Death')
plt.title('COVID-19 Stats in California')
plt.ylabel('Number of Cases in Millions')
plt.xlabel('From 01-22-20 - 12-06-20')
plt.legend()
plt.xticks([])
plt.show()
```

Conclusion:

After viewing the statistics of COVID-19 (Corona Virus) in California from 11-14-20 to 12-06-20. There is multiple points on both the "positive" and "death" flatten on the line plot. The possible explaination that could coincide with this display is the active practice of social distancing and self-quarantine contributing to "Flattening the Curve". After a amount of time of the line being flat, then it starts to rise again. Possibly due to seeing family during the holidays and accumalting isolation fatigue.

Collecting COVID-19 data from IL (Illinois).

```python
df_IL = df[df.state == 'IL']
df_IL
```

Plotting the Positive statistics on the line graph. Revealing a large increase in cases later in the year.

```python
plt.plot(df_IL['date'], df_IL['positive'], color='red')
plt.title('COVID-19 Stats in Illinois')
plt.ylabel('Number of Cases')
plt.xlabel('From 03-04-20 - 12-06-20')
plt.xticks([])
plt.show()
```

Plotting the Death statistics line graph. Showing a trend through out the year deaths line flattens, but continue to increase.

```python
plt.plot(df_IL['date'], df_IL['death'], color='black')
plt.title('COVID-19 Stats in Illinois')
plt.ylabel('Number of Deaths')
plt.xlabel('From 01-22-20 - 12-06-20')
plt.xticks([])
plt.show()
```

Plotting both lines on the same graph.

```python
plt.plot(df_IL['date'], df_IL['positive'], color='red', label='Positive')
plt.plot(df_IL['date'], df_IL['death'], color='black', label='Death')
plt.title('COVID-19 Stats in Illinois')
plt.ylabel('Number of Cases in Millions')
plt.xlabel('From 01-22-20 - 12-06-20')
plt.legend()
plt.xticks([])
plt.show()
```

Now plotting the Positive cases from California and Illinois on one graph to compare the two states.

```python
plt.plot(df_IL['date'], df_IL['positive'], color='red', label='Illinois')
plt.plot(df_CA['date'], df_CA['positive'], color='blue', label='California')
plt.title('Corona Virus cases in IL and CA')
plt.legend()
plt.xlabel('01-22-20 to 12-06-20')
plt.ylabel('Cases in the Millions')
plt.xticks([])
```

And finally, plotting the amount of deaths in each state on one line graph.

```python
plt.plot(df_IL['date'], df_IL['death'], color='red', label='Illinois')
plt.plot(df_CA['date'], df_CA['death'], color ='blue', label='California')
plt.title('Amount of COVID-19 Deaths in IL and CA')
plt.xlabel('From 1-22-20 to 12-06-20')
plt.ylabel('Deaths')
plt.legend()
plt.xticks([])
```

It is interesting that Illinois starts off with a higher death rate, but as time moves on California over takes the amount of COVID-19 deaths. This could be contributed to California getting a early grip on the situation while the state of Illinois did not take actions, Chicago implemented stay at home orders and mask mandates. It should also be noted that the population for both states are drastically different: California with 39.78 million and Illinois with 12.63 million according to the 2020 census.

Second, I will use the COVID-19 data in the United States plot out the total positive cases, deaths, and visualize them. Below I am placing the CSV file into a DataFrame.

```python
df_US = pd.read_csv('/kaggle/input/covid19-in-usa/us_covid19_daily.csv')
df_US
```

Sorting the entire DataFrame by "date" from earliest to present.

```python
df_US.sort_values(by ='date', inplace = True)
df_US
```

Plotting the "positive" numbers on a line graph.

```python
plt.plot(df_US['date'], df_US['positive'], color='red')
plt.title('COVID-19 cases in America')
plt.xlabel('From 01-22-20 to 12-06-20')
plt.ylabel('Cases in the Millions')
plt.xticks([])
plt.show()
```

Plotting the "Death" statistics on a graph.

```python
plt.plot(df_US['date'], df_US['death'], color='black')
plt.title('COVID-19 Deaths in America')
plt.xlabel('From 01-22-20 to 12-06-20')
plt.ylabel('Deaths')
plt.xticks([])
plt.show()
```

In the entire data there is a low death rate. Then a rises throughout the year. These numbers could be higher, because it is not factoring the COVID-19 death in proximity. Which means Even though there was not a death by the infection of COVID-19, but with the emergency rooms overflowing with potential COVID-19 cases. A individual could have gone to the hosiptal for a life threatening injury, but since the staff was overwhelmed with tending to the potential COVID-19 case, their attention was never given to the individual, resulting in death.
